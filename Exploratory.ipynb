# Exploring dataset



```r
df = read.csv("ool_pds.csv")
```

```r
dim(df)
```

* There are 2,294 observations and 436 variables.

## Case ID

* The first variable is the case id. It is a sequential integer. 
* No missing value.

```r
summary(df['CASEID'])
str(df['CASEID'])
sum(is.na(df['CASEID']))
c(min(df['CASEID']), max(df['CASEID']))
```

## Case ID (Wave 1)

* Wave 1 case id is a sequential integer.
* All observations have wave 1 case id.
* Case ID are not sequential

```r
str(df['W1_CASEID'])
sum(is.na(df['W1_CASEID']))
c(min(df['W1_CASEID']), max(df['W1_CASEID']))
```

## Case ID (Wave 2)

* Wave 2 case id is also an integer.
* There are 693 missing values.

```r
str(df['W2_CASEID2'])
sum(is.na(df['W2_CASEID2']))
```

### Wave 1

* There are 276 variables (out of total of 436 variables) are related to Wave 1.

```r
w1_vars = names(df)[grepl('^W1_', names(df))]
length(w1_vars)
w1_vars
```

### Wave 2

* There are 135 variables (out of total of 436 variables) are related to Wave 2.

```r
w2_vars = names(df)[grepl('^W2_', names(df))]
length(w2_vars)
w2_vars
```

```r
library(gmodels)
x = factor(df$PPETHM, labels = c("White", "Black", "Non-Hispanic", "Hispanic", "Multi-race"))
str(x)

y = factor(df$W1_A5A, labels = c("Refused", "John McCain", "Barack Obama", "Other"))
str(y)
CrossTable(x, y, dnn = c("Race", "Candidate")) 
CrossTable(df$PPETHM, df$W1_A5A, dnn = c("Race", "Candidate")) 
```

```r
x = factor(df$PPINCIMP)
str(x)

y = factor(df$W1_A5A, labels = c("Refused", "John McCain", "Barack Obama", "Other"))
str(y)
CrossTable(x, y, dnn = c("Household Income", "Candidate")) 
```
